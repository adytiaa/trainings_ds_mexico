## William Playfair (1759–1823)
* Scottish engineer, political economist, and secret agent
* Founder of graphical methods of statistics
* Invented the following charts:
  * line
  * bar
  * area
  * pie
  * circle graph
* Draftsman and personal assistant to James Watt (steam engine) in 1777
* Failed with several businesses 1781-1787
* First publication on economics 1785
* Fled to France 1787
* Took part in the Storming of the Bastille 1789
* Was part of the Scioto Land sale to French settlers 1787-1791 (sold land they did not own)
* Forced to leave France 1793
* Worked as a British secret agent during the war with France in the 1790s
* Lived in London, traveled occasionally to America, and published several books 1793-1814
* Attempted to blackmail a Scottish lord in 1816
* died in poverty

## Basic Chart Typ

### Bar Chart
* Most common chart type
* Compare items/categories that have one or multiple numerical variables
* Think about sorting the values (start with the highest numerical value on top and sort descending)
* Vertical bar charts (also *column* charts) can be used to compare items over time but should not be used for too many items
* For multiple numerical variables:
  * stacked bar chart
  * side-by-side bar chart (bars with different colors next to each other)


**Horizontal Bar Chart**

**Vertical or Column Bar Chart**

**Side-by-side Bar Chart**

### Histogram
* Shows distribution of numerical data
* Continuous data is grouped into bins
* Only a single numerical value as input
* Number of bins is defined by the user
* Often equal size bins (but not required)
* Similar to bar chart but histograms do not have categories

![histogram](img/chart_types/histogram.png)

### Box Plots
* Shows distribution of numerical data
* Statistical measures:
  * box are the values in the upper and lower quartile
  * separator in the box is the median
  * whiskers represent the distances between the lowest value to the first quartile and the fourth quartile to the highest value
* outliers may be included as individual points
* Boxplots do not show the number of values

![box_plot](img/chart_types/box_plot.png)

### Violin Plots
* Similar to box plots but also show the probability density of the data

![violin_plot](img/chart_types/violin_plot.png)

### Line Chart
* One or multiple lines that develop over time
* Continuous data
* Visualizing trends
* Chronological / time series data on x-axis

![line](img/chart_types/line.png)

### Area Chart
* Similar to line chart
* The normal area chart is confusing and therefore not used often: use a line chart instead
* Visualize parts of a whole that change over time with a **stacked area chart** or a **100% stacked area chart**

#### Isaac Newton (1643-1727)
* English mathematician, physicist, astronomer, theologian, and author
* *"one of the most influencial scientists of all time"*
* "observation that a prism separates white light into the colours of the visible spectrum"

**Dual Prism Experiment, 1671**
* white light passes through the first prism
* the first prism breaks the light into its spectral colors
* only one color passes the second prism (here red)
* the color does not change when passing the second prism (it stays red)
* conclusion: white light is a mixture of different colors and the prism does not affect the color

#### Color Harmonies
Asthetically pleasing color combinations.

##### Monochromatic
* Only one color (single hue)
* Adding different tones, tints, and shades

![red_monochrome](img/red_monochrome.svg)
https://upload.wikimedia.org/wikipedia/commons/1/1a/CPT-Websites-monochrome.svg

##### Analogous
* Adjacent colors on wheel
* Pleasing to the eye
* Works well with contrast
* One color dominates, a second supports and the third color accents

##### Complementary
* Opposing colors on the wheel
* Use one color predominantly (normally choose the less dominant one)

##### Split Complementary
* One complimentary end is extended
* Same contrast as complementary cholor scheme but less tension

##### Triadic
* Equal distance on wheel
* One color should dominate and the other two are used as accent

##### Tetradic (Double Complementary or Rectengular)
* Two pairs of opposing colors
* Best used for foreground / background
* Do not use 25% each

##### Square
* Similar to tetradic but even distance between colors
* Most possible color combinations

### Color Models
* Abstract mathematical model
* Colors as tuples of numbers (usually 3-4 values)
* Color model + mapping function = **color space**
* A color space allows analog and digital reproducible representations of color

* **Additive Color Models:**
  * start with darkness
  * mix light colors together to create new colors
  * e.g. RGB color model

* **Subtractive Color Models:**
  * start with light
  * create new colors by absorbing parts of the spectrum of light
  * like mixing paints
  * e.g. CMY(K) color model


#### RGB Color Model
* Additive color model with red (R), green (G), and blue (B) as primary colors
* Main purpose: display of images in electronic systems

![rgb](img/rgb.svg)
https://upload.wikimedia.org/wikipedia/commons/c/c2/AdditiveColor.svg

#### CYM(K) Color Model
* Substractive color model with cyan (C), yellow (Y), magenta (M) as primary colors
* Additional black (K) for dark colors
* Used in printing

![cmy](img/cmy.svg)
https://upload.wikimedia.org/wikipedia/commons/1/19/SubtractiveColor.svg

#### HSL & HSV Color Models
* HSL: hue, saturation, lightness
* HSV: hue, saturation, value
* "Geometric transformations of the RGB cube into cylindrical form"
* Developed in the 1970s
* Additive primary and secondary (pure) colors: red, yellow, green, cyan, blue, and magenta

**Hue**
* Color on the color wheel
* Specified as angle
* 0° and 360°: red
* 120°: green
* 240°: blue

![hue](img/hue.svg)
https://upload.wikimedia.org/wikipedia/commons/a/ad/HueScale.svg

**Saturation** 
* intensity of the color
* combination of white light
* 0% at the left is neutral grey (or black or white)
* 100% at the right is the pure color

![saturation_red](img/saturation_red.png)
https://upload.wikimedia.org/wikipedia/commons/5/5e/Saturation_horizontal.png

**Chroma**
* perceived strength of a color
* "describes how different from a grey of the same lightness such an object color appears to be"
* "**colorfulness** of an object relative to the brightness of a white object similarly illuminated"
* "open-ended scale of perceptually uniform steps"
* 0% full white
* 100% full black or full color

http://www.huevaluechroma.com/082.php and https://en.wikipedia.org/wiki/Colorfulness

* 100% full white

**Value**
* 0% nearly black (no brightness)

##### CIELAB Color Space
* Also *CIE L&ast;a&ast;b* or simply *Lab*
* Based on Munsell color system
* Developed after human vision
* Device independent
* License-free and public domain
* Numerical change in values = perceived change
* Colors are calculated relative to a given white point
* Hue, chroma, lightness
* 3 components:
  * Lightness (0-100%)
  * Two bipolar color components
    * a (yellow-blue)
    * b (green-red/magenta)
* Contains more colors than RGB
* HCL: cylindrical transformation of CIELAB

#### Different Colors for Different Data 
* **Sequential / diverging data**
  * Gradients
  * Uniform variation in lightness and saturation
  * Simultaneous shift in hue is preferable
  
![three_perceptual_palettes](img/three_perceptual_palettes.png)
https://earthobservatory.nasa.gov/blogs/elegantfigures/wp-content/uploads/sites/4/2013/08/three_perceptual_palettes_618.png

* **Divergent (bipolar) data**
  * Central value (breakpoint) with neutral color (white or light grey)
  * Different hues that vary from center

![divergent_palettes](img/divergent_palettes.png)
https://earthobservatory.nasa.gov/blogs/elegantfigures/wp-content/uploads/sites/4/2013/08/divergent_palettes_618.png

* **Qualitative / categorical data**
  * Set of opposing colors
  * As distinct colors as possible
  * Perception limits the maximum number of categories to 12 (practially even fewer)
  * If possible, group similar classes (e.g. shades of green)
  * Combine colors with shapes for many categories (see "Shape" chapter below(
